# Forecasting Total Cases Worldwide

```python
## Prophet, a procedure for forecasting time-series data based on additive model where non linear trends are fit with yearly,
## weekly and daily seasonality.It is robust to outliers and missing data
## in prophet we need to give 2 inputs as it is a dataframe with 2 columns
```

```python
from fbprophet import Prophet
```

```python
confirmed = df.groupby('Date').sum()['Confirmed'].reset_index()
deaths = df.groupby('Date').sum()['Deaths'].reset_index()
recovered = df.groupby('Date').sum()['Recovered'].reset_index()
```

```python
## 2 columns = ds(datestamp) and y is values column (which is a numerical value)
```

```python
confirmed.columns = ['ds','y']
#confirmed['ds'] = confirmed['ds'].dt.date
confirmed['ds'] = pd.to_datetime(confirmed['ds'])
```

```python
confirmed.tail()
```

# COVID Confirmed Cases Worldwide with Prophet

```python
## Generating a week ahead forecast using Prophet, with 95% prediction interval by creating a base model with no 
## seasonality parameters (as Corona virus is not having any seasonality) and additional regressors
```

```python
m = Prophet(interval_width = 0.95)  ## interval_width = confidence
m.fit(confirmed)
future = m.make_future_dataframe(periods=7) ## 7 days period
future.tail()
```

```python
## The predict method will assign each row in future a predicted value which it names yhat.
```

```python
## predicting future with date , and the upper and the lower limit of of y values
```

```python
forecast = m.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']]
```

```python
## plotting the forecast by calling Prophet.plot method
confirmed_forecast_plot = m.plot(forecast)
```

```python
## The Lockdown on Tuesday lead to the dip in the curve
confirmed_forecast_plot = m.plot_components(forecast)
```

# COVID Death Cases Worldwide with Prophet

```python
deaths.columns = ['ds','y']
deaths['ds'] = pd.to_datetime(deaths['ds'])
```

```python
m = Prophet(interval_width = 0.95)  ## interval_width = confidence
m.fit(deaths)
future = m.make_future_dataframe(periods=7) ## 7 days period
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
deaths_forecast_plot = m.plot(forecast)
## the markers are original values and the blue line is the forecast value
```

```python
deaths_forecast_plot = m.plot_components(forecast)
```
